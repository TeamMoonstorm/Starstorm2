```python
import os
```

```python
file_to_clean = "./SS2-Project/Assets/Starstorm2/AssetStorm/Scenes/VoidShop.unity"
META_FILE_EXT  = ".meta"
```

```python
# recursively iterate through all .meta files in the repo and load the txt into a list of strings
def load_meta_files():
    meta_files = []
    for root, dirs, files in os.walk("./SS2-Project"):
        for file in files:
            if file.endswith(META_FILE_EXT):
                with open(os.path.join(root, file), "r") as f:
                    meta_files.append(f.read())
    return meta_files
```

```python
meta_files = load_meta_files()
```

```python
meta_files[0]
```

```python
# Next we go through each meta file and create a dictionary of guids
# this will be used to confirm whether a guid exists or not late
def create_meta_dict(meta_files: list[str]) -> dict:
    guid_dict = {}
    for meta in meta_files:
        try:
            # split the meta file into lines
            lines = meta.split("\n")

            # guid is found on the 2nd line in the file
            guid = lines[1].split("guid:")[1].strip()

            if guid not in guid_dict:
                guid_dict[guid] = True
            else:
                print(f"Duplicate GUID found: {guid}")
        except Exception as e:
            print(f"Error parsing meta file")
            
    return guid_dict
```

```python
guid_dict = create_meta_dict(meta_files)
```

```python
import re

# Sample lines
lines = [
    "{fileID: 2100000, guid: 90d4220f24fca754ca88bb942be70dde, type: 2}",
    "skyboxMaterial: {fileID: 2100000, guid: d8f4212758895924e880b8d2488679e8, type: 2}"
]

# Regular expression to match GUIDs
guid_pattern = re.compile(r'guid: ([a-f0-9]{32})')

# Extract GUIDs
guids = [guid_pattern.search(line).group(1) for line in lines if guid_pattern.search(line)]

# Print the extracted GUIDs
for guid in guids:
    print(guid)

```

```python
found_guids = []
missing_guids = []

# Regular expression to match GUIDs
guid_pattern = re.compile(r'guid: ([a-f0-9]{32})')

for idx, object in enumerate(objects):
    try:
        # split the object into lines
        lines = object.split("\n")

        guids = [guid_pattern.search(line).group(1) for line in lines if guid_pattern.search(line)]

        for guid in guids:
            if guid in guid_dict:
                found_guids.append(guid)
            else:
                missing_guids.append(guid)
                
    except Exception as e:
        print(f"Error parsing object")
```
